# ML для разработчиков: от практики к теории и обратно

## Чему мы научимся:
* Рассмотрим главные подходы в ML
* Разберемся с математикой и популярными инструментами
* Научимся решать популярные проблемы предсказывания будущих продаж, распознавания объектов на фото, рекомендации фильмов, генерацию человекоподобного текста и другие
* Все проблемы будем решать как с помощью распространенных библиотек так с помощью собственного кода без использования специальных библиотек
* Разберемся в главных мифах про ML - про математику, специальное образование и научный подход

## Наш подход:

Основной подход курса лежит между:
* быстрым стартом использования специальных библиотек для решения самых современных задач без понимания того, как они работают 

И 

* супер-теоретическим подходом, в котором сперва проходится полностью математический аппарат, рассматривается история  подходов для работы с неструктурированными данными и только потом появляется понимание зачем это нужно

Мы делимся курсом ML для разработчиков - смотрим на современные алгоритмы машинного обучения с точки зрения кода - и  как разработчики получаем понимание как все устроено и как работают основные концепции на которых строится современная наука. Мы не будем делать математических допущений и упрощений - будем решать реальные задачи с хорошим качеством.

Все задачи мы будем решать как с помощью стандартных библиотек - так и с помощью кода, который будем писать сами без использования специальных ml-библиотек.

В конце каждой проблемы будет предложено решить несколько задач на kaggle - ресурсе, на котором соревнуются лучшие специалисты со всего мира.
